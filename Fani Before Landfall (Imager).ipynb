```python
import h5py
import os
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

path='3DIMG_15JUN2022_1000_L1B_STD.h5'
fanibl=h5py.File(r'C:\Users\HP\OneDrive\Desktop\HD5 Collection\3DIMG_03MAY2019_0230_L1B_STD.h5')
```

```python
#All details about paylod are taken from INSAT_3D catalog 
#Link - https://www.researchgate.net/publication/323401247_INSAT-3D_DATA_PRODUCTS_CATALOG#fullTextFileContent
#First relevant HD5 file is loaded, then for each imager channel, we plot a global and an asia-specific plot.
#For each channel, the channel produced numpy array is loaded and plotted as shown below.
```

```python
viscount3D=fanibl['IMG_VIS'][:]
viscount=np.squeeze(viscount3D)

fig,(ax1,ax2)=plt.subplots(nrows=1,ncols=2,figsize=(15,15)) #Defining the figure and axes

image_viscount_global=ax1.imshow(viscount,label="Global") #First image
plt.colorbar(image_viscount_global,ax=ax1,fraction=0.046, pad=0.04) #Colorbar for first image

image_viscount_asia=ax2.imshow(viscount,label="Asia specific") #Second image
plt.axis([3000, 7000, 6000, 2000])  #Axis restriction for asia-specific plot.
plt.colorbar(image_viscount_asia,fraction=0.046,ax=ax2, pad=0.04) #Colorbar for 2nd subplot

ax1.title.set_text('Global plot') #1st subplot title
ax2.title.set_text('Asia-specific plot') #2nd subplot title
ax1.grid()
ax2.grid()

fig.suptitle('Visible Count for Fani before landfall,λ=0.55–0.75 μm,GR=1KM,IGFOV=28μrad, t=0730-0800 IST 03/05/2019, L1B INSAT-3D Imager',y=0.70)
```
